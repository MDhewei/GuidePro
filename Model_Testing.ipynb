## Compare the performance of different methods on dependent datasets

```python
import pandas as pd
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import scale
from scipy.stats import spearmanr
from sklearn.linear_model import Ridge
import random
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt 
import seaborn as sns

## Split the training set for sub-predictor training and assembly training
def SplitData(df_train):
    df_sep = DataFrame()
    df_asm = DataFrame()
    ## For each gene randomly selet 100 sgRNA and split into 50/50
    for gene in list(set(df_train['Symbol'])):
        df_gene = df_train[df_train['Symbol']==gene]
        df_gene.index = range(df_gene.shape[0])
        size = df_gene.shape[0]
        if size < 100: ## If one gene got less than 100 sgRNA, choose all for training
            ix_ls = random.sample(range(size),size)
            ix_sep = ix_ls[0:]
            ix_asm = ix_ls[int(size/2):]
        else: 
            ix_ls = random.sample(range(size),100)
            ix_sep = ix_ls[0:]
            ix_asm = ix_ls[50:]
    
        df_sep = pd.concat([df_sep,df_gene.iloc[ix_sep,:]])
        df_asm = pd.concat([df_asm,df_gene.iloc[ix_asm,:]])
    return df_sep,df_asm    
        

def Assembly(df_sep,df_asm,df_test):
    
    s_seq = df_sep.loc[:,['Essential_Zscore','Doench_score','SSC_score','DeepHF_score']]
    
    a_seq = df_asm.loc[:,['Essential_Zscore','Doench_score','SSC_score','DeepHF_score']]
    
    s_pro = df_sep.loc[:,['Essential_Zscore','Domain_pfam','PROVEAN_score','SIFT_score',
                           'Helix','Sheet','Phosphorylation','Acetylation']]
    
    a_pro = df_asm.loc[:,['Essential_Zscore','Domain_pfam','PROVEAN_score','SIFT_score',
                           'Helix','Sheet','Phosphorylation','Acetylation']]
    
    s_ind = df_sep.loc[:,['Essential_Zscore','FrameShift_IndelPhi','FrameShift_Lindel']]
    
    a_ind = df_asm.loc[:,['Essential_Zscore','FrameShift_IndelPhi','FrameShift_Lindel']]
    
    
    svr1 = svm.SVR(kernel='rbf',C=1,gamma=0.01) #RandomForestRegressor()#svm.SVR(kernel='rbf')
    svr2 = svm.SVR(kernel='rbf',C=1,gamma=0.01) #RandomForestRegressor()#svm.SVR(kernel='rbf')
    svr3 = svm.SVR(kernel='rbf',C=1,gamma=0.01) #RandomForestRegressor()#svm.SVR(kernel='rbf')
    
    sv_seq = svr1.fit(scale(s_seq.iloc[:,1:]),s_seq['Essential_Zscore'])
    sv_pro = svr2.fit(scale(s_pro.iloc[:,1:]),s_pro['Essential_Zscore'])
    sv_ind = svr3.fit(scale(s_ind.iloc[:,1:]),s_ind['Essential_Zscore'])
    
    df_asm['seq_score'] = list(sv_seq.predict(scale(a_seq.iloc[:,1:])))
    df_asm['pro_score'] = list(sv_pro.predict(scale(a_pro.iloc[:,1:])))
    df_asm['ind_score'] = list(sv_ind.predict(scale(a_ind.iloc[:,1:])))
    
    df_asm['seq_mod'] = df_asm['seq_score']*df_asm['PROVEAN_STD']
    df_asm['pro_mod'] = df_asm['pro_score']*df_asm['PROVEAN_STD']
    df_asm['ind_mod'] = df_asm['ind_score']*df_asm['PROVEAN_STD']
    
    svl = LinearRegression() 
    asm = svl.fit(scale(df_asm.loc[:,['seq_score','pro_score','ind_score']]),df_asm['Essential_Zscore'])

    seq_score = list(sv_seq.predict(scale(df_test.loc[:,['doench_score','SSC_score','DeepHF_score']])))
    pro_score = list(sv_pro.predict(scale(df_test.loc[:,['Domain_pfam','PROVEAN_score','SIFT_score','Helix','Sheet','Phosphorylation','Acetylation']])))
    ind_score = list(sv_ind.predict(scale(df_test.loc[:,['FrameShift_IndelPhi','FrameShift_Lindel']])))
    
    df_test['seq_score'] = seq_score
    df_test['pro_score'] = pro_score
    df_test['ind_score'] = ind_score
    
    combined_score = list(asm.predict(scale(df_test.loc[:,['seq_score','pro_score','ind_score']])))
    
    return seq_score,pro_score,ind_score,combined_score


All_dic = {}

testset = ['xu2015_HL60','xu2015_KBM7','chari2015','hart2016-Rpe1','hart2016-Hct1162', 
          'hart2016-HelaLib1', 'hart2016-HelaLib2','doench2014']

df_train = pd.read_csv('./Training_datasets/Tiling_training_v2.csv')


df_result = DataFrame()

for data in testset:
    print('Testing on %s!'%data)
    corr_dic = {}
    df_test = pd.read_csv('./Dependent_datasets/Testing_'+data+'.csv')
    
    df_test['CRISPRscan.score'] = df_test['crisprScan']
    df_test['Wang.score'] = df_test['wangOrig']
    df_test['Chari.score'] = df_test['chariRaw']
    df_test['WuCRISPR.score'] = df_test['wuCrispr']
    
    df_tr = df_train.loc[:,['Symbol','Essential_Zscore','Doench_score','SSC_score','DeepHF_score',
                           'FrameShift_IndelPhi','FrameShift_Lindel','FrameShift_FORECasT','Domain_pfam','PROVEAN_score',
                           'SIFT_score','Helix','Sheet','Phosphorylation','Acetylation','PROVEAN_STD','SIFT_STD']]

    
    df_te = df_test.loc[:,['dataset','sgRNASeq','Symbol','Essential_Zscore','doench_score','SSC_score','DeepHF_score','DeepCRISPR_score',
                           'FrameShift_IndelPhi','FrameShift_Lindel','FrameShift_FORECasT','Domain_pfam','PROVEAN_score',
                           'SIFT_score','Helix','Sheet','Phosphorylation','Acetylation','PROVEAN_STD',
                           'CRISPRscan.score','Wang.score','Chari.score','WuCRISPR.score','CRISPRO_score']]
    
    seq_ls = []; pro_ls = []; ind_ls = []; com_normal = []; 
    for n in range(10):
        df_sep,df_asm = SplitData(df_tr)
        seq_score,pro_score,ind_score,combined1 = Assembly(df_sep,df_asm,df_te)
        seq_ls.append(seq_score); pro_ls.append(pro_score);ind_ls.append(ind_score)
        com_normal.append(combined1)#;com_adjust.append(combined2)
    
    
    df_te['LoF_seq'] = list(DataFrame(seq_ls).T.mean(axis=1))
    df_te['LoF_pro'] = list(DataFrame(pro_ls).T.mean(axis=1))
    df_te['LoF_ind'] = list(DataFrame(ind_ls).T.mean(axis=1))
    df_te['Combined'] = list(DataFrame(com_normal).T.mean(axis=1))
    
    df_result = pd.concat([df_result,df_te])
    
    corr_dic['CRISPRscan'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['CRISPRscan.score'])[0])
    corr_dic['Wang.score'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['Wang.score'])[0])
    corr_dic['Chari.score'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['Chari.score'])[0])
    corr_dic['WuCRISPR'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['WuCRISPR.score'])[0])
    corr_dic['Doench.score'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['doench_score'])[0])
    corr_dic['SSC.score'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['SSC_score'])[0])
    corr_dic['DeepCRISPR'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['DeepCRISPR_score'])[0])
    corr_dic['DeepHF'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['DeepHF_score'])[0])
    corr_dic['CRISPRO'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['CRISPRO_score'])[0])
    corr_dic['SVM_seq'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['LoF_seq'])[0])
    corr_dic['SVM_ind'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['LoF_ind'])[0])
    corr_dic['SVM_pro'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['LoF_pro'])[0])
    corr_dic['Combined'] = abs(spearmanr(df_test['Essential_Zscore'],df_te['Combined'])[0])
    All_dic[data] = corr_dic

df_result.to_csv('Dependent_test_result.csv',index=False)

df = DataFrame(All_dic,columns=testset,index=['CRISPRscan','Wang.score','Chari.score','WuCRISPR','Doench.score',
                        'SSC.score','DeepCRISPR','DeepHF','CRISPRO','SVM_seq','SVM_pro','SVM_ind','Combined'])

df['Average'] = df.mean(axis=1)
plt.figure(figsize=(7,7),dpi=300)
plt.rcParams['font.family'] = 'Helvetica'
sns.heatmap(df,annot=True, fmt=".3f")
plt.show()
```
