```python
from __future__ import division
import json,os
import pandas as pd
from pandas import DataFrame

ProteinFilePath = '/home/data/sgEfficiencyPrediction/Protein' ## the folder containing all human protein information
exons_dic = json.loads(open(os.path.join(ProteinFilePath,'Exons_dic_hg19')).read())

alias_dic = {'PBEF1':'NAMPT','FRAP1':'MTOR','SFRS2':'SRSF2','CMYC':'MYC','MYST1':'KAT8','PRSS15':'LONP1',
             'SMC2L1':'SMC2','DDX48':'EIF43A','VPRBP':'DCAF1','CDC2':'CDK1'}


testset1 = ['xu2015_HL60','xu2015_KBM7','chari2015','hart2016-Rpe1','hart2016-Hct1162', 
           'hart2016-HelaLib1', 'hart2016-HelaLib2','doench2014','doench2016','doench2016_plx']

testset2 = ['Schoonenberg2018','Bertomeu2018','Evers2016-RT112','Evers2016-UMUC3','Sanjana2014-GeCKOv2']


for data in testset2:
    print data
    df = pd.read_csv('/home/whe/Work_progress/sgRNA_yitao/Test_Set_2/Testing_'+data+'_update.csv')
    df_new = DataFrame()
    aa_ls = []
    for i in range(df_new.shape[0]):
        gene = df_new['Symbol'][i]
        if gene in exons_dic:
            aa =  exons_dic[gene][0]['AA.num']
        else:
            aa =  exons_dic[alias_dic[gene]][0]['AA.num']
    
        aa_ls.append(aa)
    
    df_new['AA.num'] = [(aa-min(aa_ls))/(max(aa_ls)-min(aa_ls)) for aa in aa_ls]
    
    df_new.to_csv('/home/whe/Work_progress/sgRNA_yitao/Test_Set_2/Testing_'+data+'_update.csv',index=False)

```
